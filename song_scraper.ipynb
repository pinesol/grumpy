```python
def get_lyrics(artist,song):
    """ Scrapes lyrics from a songlyrics page"""
    limit_we_set = 100
    url = ('http://www.songlyrics.com/{0}/{1}-lyrics/'.format(artist,song)).lower()
    print url
    #driver = webdriver.Chrome()
    #wait = WebDriverWait(driver, 10)
    # load the page
    
    try:
        driver.set_page_load_timeout(10)
        driver.get(url)
    except Exception:
        print 'time out'
        #driver.send_keys(Keys.CONTROL +'Escape')
    
    #driver.get(url)
    #WebDriverWait(driver, 5).until(lambda d: d.find_element_by_id('songLyricsDiv').is_displayed())
    
    print 'page loaded'
    
    
    try:
        text = driver.find_element_by_id('songLyricsDiv').text
        print text
        print "Type is : " + type(text)
        print "Len is : "+ len(text)
        if len(text)>limit_we_set:        
        
            file_name = 'Lyrics/'+artist+' - '+song+'.txt'
            print 'saving file : '+'Lyrics/'+artist+' - '+song+'.txt'
            text_file = open(file_name, "w")
            text_file.write(text)
            text_file.close()
    
    except:
        pass
    
 #driver.quit()
```
